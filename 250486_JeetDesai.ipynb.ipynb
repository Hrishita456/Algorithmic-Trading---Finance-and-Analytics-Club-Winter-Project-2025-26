```python
Input = input("Enter the word: ")
Word = Input.lower()
for i in range(0,len(Word)):
  if (Word[i] != Word[len(Word)-1-i]):
    print("Your word is not a palindrome")
    break
  elif (i == len(Word)-1):
    print("Your word is a palindrome")
  else:
    continue
```
