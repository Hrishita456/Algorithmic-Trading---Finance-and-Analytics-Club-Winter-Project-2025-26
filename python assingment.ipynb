```python
# Question 1
list=[1,56,8,35,89,43,78,446,88,23,487,569,8,9,6,55]
max=list[0]
min=list[0]
for i in range(len(list)):

  if list[i]>max:
    max=list[i]
  elif list[i]<min:
    min=list[i]
print(max)
print(min)

```

```python
# Question 2
s=input('enter a string')
l=len(s)
a=''
for i in range(l):
  a=a+s[l-i-1]
if a==s:
    print('it is a palindrome')
```

```python
#Question 3

# let the value of x be 54

lt=[1,56,60,35,89,43,78,446,88,23,487,569,8,9,6,55]
lt.sort()
for i in range(len(lt)):
  for j in range(len(lt)):
    if (lt[i]-lt[j]==54):
      print('the pair exist and the pair is')
      print(lt[i],lt[j])
```
