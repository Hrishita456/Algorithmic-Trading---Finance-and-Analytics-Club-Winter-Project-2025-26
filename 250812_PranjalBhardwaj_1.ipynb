```python
#question 1
n=int(input("Enter number of elements:"))
arr=[]
for i in range(n):
  val=int(input("Enter element:"))
  arr.append(val)

max=arr[0]
min=arr[0]
for x in arr:
  if x<min:
    min=x
  if x>max:
    max=x

print("Maximum",max)
print("Minimum",min)
```

```python
#question 2
s=input("Enter string:")
a=0
b=len(s)-1
is_palindrome = True
while a<b:
  if s[a]!=s[b]:
    is_palindrome = False
    break
  a+=1
  b-=1
if is_palindrome:
  print("Palindrome")
else:
  print("Not a palindrome")
```

```python
#question 3
n=int(input("Enter number of elements:"))
arr=[]
for i in range(n):
  val=int(input("Enter an element:"))
  arr.append(val)

x=int(input("Enter absolute difference:"))
arr.sort()
i=0
j=1

while i<n and j<n:
  abs_diff=arr[j]-arr[i]
  if abs_diff==x and i!=j:
    print("Pair found:",arr[i],arr[j])
    break
  elif abs_diff<x:
    j+=1
  else:
    i+=1
```

```python
#question 4

n =int(input("Enter number of elements: "))

arr=[]
for i in range(n):
    val=int(input("Enter element: "))
    arr.append(val)

a = int(input("Enter value a: "))
b = int(input("Enter value b: "))
c = int(input("Enter value c: "))

count_a = 0
count_b = 0
count_c = 0

for x in arr:
    if x == a:
        count_a = count_a + 1
    elif x == b:
        count_b = count_b + 1
    else:
        count_c = count_c + 1

small = a
mid = b
large = c

if small > mid:
    temp = small
    small = mid
    mid = temp

if mid > large:
    temp = mid
    mid = large
    large = temp

if small > mid:
    temp = small
    small = mid
    mid = temp

sorted_arr = []

for i in range(count_a):
    sorted_arr.append(a if a == small else (b if b == small else c))

for i in range(count_b):
    sorted_arr.append(a if a == mid else (b if b == mid else c))

for i in range(count_c):
    sorted_arr.append(a if a == large else (b if b == large else c))

print("Sorted array:",sorted_arr)
```
