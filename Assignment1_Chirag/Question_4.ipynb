Question_4. You are given an array of length n, and distinct integers a, b, and c. All elements of the array
are one of a, b, or c. You are required to sort the array without the use of a conventional
sorting algorithm.


```python
import numpy as np
```

```python
# Using Sorting
# Given an array of size n consisting of distinct integers let them be a<b<c
def Sort(arr, a, b, c):
  if a > b:             # Sorting in ascending order a < b < c
    a, b = b, a
  if b > c:
    b, c = c, b
  if a > b:
    a, b = b, a
  num_a, num_b, num_c = 0, 0, 0
  for num in arr:
    if num == a:
      num_a += 1
    elif num == b:
      num_b += 1
    else:
      num_c += 1
  sorted_arr = []
  for i in range(num_a):
    sorted_arr.append(a)
  for i in range(num_b):
    sorted_arr.append(b)
  for i in range(num_c):
    sorted_arr.append(c)
  return sorted_arr
size_arr = int(input("Enter the size of array:"))
arr = np.random.randint(3, size =(size_arr) )
Sort(arr, 2, 1, 0)
```

```python
# If the array is taken as input by user
size_arr = int(input("Enter the size of array: "))
a = int(input("Enter the value of a: "))
b = int(input("Enter the value of b: "))
c = int(input("Enter the value of c: "))
arr = []
for i in range(size_arr):
  arr.append(int(input(f"Enter (a/b/c) number{i} : ")))
Sort(arr, a, b, c)

```

```python
# Without Conventional Sorting by three pointer method
def Sort(arr, a, b, c):
  if a > b:             # Sorting in ascending order a < b < c
    a, b = b, a
  if b > c:
    b, c = c, b
  if a > b:
    a, b = b, a
  low, mid, high = 0, 0 , len(arr) - 1
  while mid <= high:
    if arr[mid] == a:
      arr[low],arr[mid] = arr[mid],arr[low]    # Swap the low and mid values
      low += 1
      mid += 1
    elif arr[mid] == b:
      mid += 1
    else:
      arr[mid],arr[high] = arr[high],arr[mid]
      high -= 1
  return arr

# Given an array
size_arr = int(input("Enter the size of array:"))
arr = np.random.randint(3, size =(size_arr) )
Sort(arr, 2, 1, 0)
```
