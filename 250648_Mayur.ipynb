```python
#Question 1

array = [int(x) for x in input().split()]

max_val = array[0]
min_val = array[0]

for number in array:

  if number > max_val:
   max_val = number

  if number < min_val:
   min_val = number

print(max_val , min_val)
```

```python
#Question 2
def check_palindrome(string):
  n = len(string)

  left = 0
  right = n - 1

  while left < right:

    if string[left] != string[right]:
            return False
            break

    left += 1
    right -= 1

  return True
```

```python
#Question 3
def find_pair(array, x):
    array.sort()

    n = len(array)

    left = 0
    right = n-1

    while right > 0:

        current_diff = array[right] - array[left]

        if current_diff == x:
            return (array[left], array[right])

        elif current_diff < x:
            right -= 1

        else:
            left += 1

        if left == right:
            right -= 1

    return None
```
